```python
#Importamos el Data Frame de precipitaciones de la clase pasada
df_lluvias  = pd.read_csv(csv_files[2])
df_lluvias.head()
```

```python
#Ponemos a Year como indice del Data Frame y lo eliminamos de las columnas
df_lluvias.index = df_lluvias['Year']
```

```python
df_lluvias = df_lluvias.drop('Year', axis='columns')
```

```python
# Para el eje x, seleccionamos los años 
x = df_lluvias.index
```

```python
# Para el eje y, acumulamos las precipitaciones por año
y = df_lluvias
```

```python
df_lluvias = df_lluvias.set_index('Year')
```

```python
df_lluvias.sum(axis='columns').to_frame().T
```

```python
suma_lluvia = df_lluvias.sum(axis='columns')
```

```python
# GRAFICO DE LINEAS de estados

plt.plot(df_lluvias.index,df_lluvias.sum(axis='columns'))
```

```python
# Definimos los objetos fig y ax, los cuales contendran la figura (Grafico de Lineas de objeto)
fig, ax = plt.subplots(figsize=(6,3))
ax.plot(suma_lluvia.index,suma_lluvia.values)
```

```python
fig, ax = plt.subplots(figsize=(6,3))
ax.plot(suma_lluvia.index,suma_lluvia.values,label='Lluvias') 
ax.set_title('Lluvias ')
ax.set_xlabel('Año')
ax.set_ylabel('Lluvias (mm)')
ax.grid(True)
ax.legend()
plt.show()
```

```python
maximo_enero = df_lluvias['Jan'].max()
maximo_febrero =  df_lluvias['Feb'].max()
```

```python
fig, ax = plt.subplots(figsize=(12, 3))  
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')
ax.axhline(maximo_enero, color='red', linestyle='--', alpha=0.5,    linewidth=3, label='Máxima de enero')
ax.axhline(maximo_febrero, color='red', linestyle=':', alpha=0.5, linewidth=3, label='Máxima de febrero')
ax.legend()
```
